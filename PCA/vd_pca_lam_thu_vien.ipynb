## Ví dụ 2: PCA tính bằng thư viện

```python
import numpy as np
from sklearn.decomposition import PCA
```

```python
np.random.seed(1)
A = np.random.randint(1,256,(3000,300))
A.shape
```

```python
# create the transform
pca = PCA(n_components=3) # số cột là 3
# fit transform
pca.fit(A)
```

```python
# transform data (project data)
B = pca.transform(A) # với B là ma trận kết quả của phép chiếu
B
```

```python
# in eigen vectors
print(pca.components_.T)
```

```python
# in eigen values
print(pca.explained_variance_) # các giá trị đơn giảm dần
```

```python
# xem tỷ lệ nén
pca.explained_variance_ratio_
```

```python
sum(pca.explained_variance_ratio_)
```
